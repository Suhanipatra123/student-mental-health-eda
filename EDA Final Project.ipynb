# 🚨 Outlier Detection

```python
# Loop through all numerical columns
for col in num_cols:
    
    # Set figure size for better visualization
    plt.figure(figsize=(5,3))
    
    # Create boxplot to detect outliers
    sns.boxplot(x=df[col], color='lightblue')
    
    # Add title for each plot
    plt.title(f"Outlier Detection - {col}")
    
    # Add grid for better readability
    plt.grid(True)
    
    # Display the plot
    plt.show()
```

Outliers are extreme values that differ from the majority of data.Boxplots are used to visually identify these unusual values.Outliers can affect statistical analysis and model accuracy.Detecting them is important before further processing.

# ✂️ Outlier Treatment (IQR)

```python
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = np.where(df[col] > upper, upper,
                       np.where(df[col] < lower, lower, df[col]))
```

```python
df.shape
```

```python
for col in num_cols:
    plt.figure(figsize=(5,3))
    sns.boxplot(x=df[col], color='lightblue')
    plt.title(f"After Treatment - {col}")
    plt.grid(True)
    plt.show()
```

Outliers are treated using the IQR (Interquartile Range) method.Values beyond acceptable limits are capped to reduce their impact.This helps in stabilizing the data distribution.It improves the robustness of analysis and models.

# 🧪QQ PLOT

```python
# Import required library
import scipy.stats as stats

# Style
sns.set_style("whitegrid")

# Loop through numeric columns
for col in df.select_dtypes(include='number').columns:
    
    # Create figure (compact size)
    plt.figure(figsize=(6, 4))
    
    # Background color
    plt.gca().set_facecolor("#f7f9fb")  
    
    # QQ Plot
    stats.probplot(df[col], dist="norm", plot=plt)
    
    # Title
    plt.title(f"QQ Plot - {col}", fontsize=11, fontweight='bold')
    
    # Clean layout
    plt.tight_layout()
    
    # Save image
    plt.savefig(f"qqplot_{col}.png", dpi=100, bbox_inches='tight')
    
    # Show plot
    plt.show()
```

QQ plot checks if data follows a normal distribution.It compares actual data with theoretical distribution.Points along a straight line indicate normality.This helps in understanding data distribution assumptions.
